Graficar los valores globales y los de Mexico para: 'Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita'

```python
def getDataCountry(country = "Mexico", dataCO2 = pd.read_csv('GCB2022v27_MtCO2_flat.csv')):
    dataCountry = dataCO2[dataCO2["Country"] == country]
    columns = dataCountry.columns
    year = dataCountry["Year"].values
    data = dataCountry[columns[3:]].set_index(year)
    return data
```

```python
dataMEX    = getDataCountry()
dataGlobal = getDataCountry("Global") 
```

```python
dataMEX.columns
```

```python
for column in dataMEX.columns:
    fig , ax= plt.subplots()
    ax.plot(dataMEX[column], label = "Mexico")
    ax.plot(dataGlobal[column], label = "Global")
    ax.set_title(column)
    ax.legend()
    plt.show(fig)
    plt.close(fig)
```

Extraer datos de los paÃ­ses pertenecientes al G20 y graficar las columnas del punto anterior.

```python
G20 = ["Argentina", "Australia", "Brazil", "Canada", "China", "France", "Germany", "India", "Indonesia", "Italy", "Japan", "South Korea", "Mexico", "Russia", "Saudi Arabia", "South Africa", "Turkey", "United Kingdom", "USA"]
```

```python
for column in dataMEX.columns:
    fig , ax= plt.subplots(figsize=(10, 7))
    
    for country in G20:   
        dataCountry = getDataCountry(country)[column]
        ax.plot(dataCountry, label = country)
        
    ax.set_title(column)
    ax.legend()
    plt.show(fig)
    plt.close(fig)
```

Â¿CuÃ¡les son los 3 paÃ­ses que mÃ¡s contaminan en cada uno de los casos?

```python
data = dataCO2.groupby(["Country"]).sum().drop(columns=["Year"]).loc[G20]
for column in data.columns:
    print(f"PaÃ­ses que mÃ¡s contaminan en la categorÃ­a <<{column}>>")
    print(data.sort_values(by=[column], ascending = False)[0:3][column])
    print()
```
