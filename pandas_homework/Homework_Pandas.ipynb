Graficar los valores globales y los de Mexico para: 'Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita'

```python
def getDataCountry(country = "Mexico", dataCO2 = pd.read_csv('GCB2022v27_MtCO2_flat.csv')):
    
    dataCountry = dataCO2[dataCO2["Country"] == country]
    columns = dataCountry.columns
    year = dataCountry["Year"].values
    data = dataCountry[columns[3:]].set_index(year)
    
    return data[data["Total"] != 0.0]
```

```python
dataMEX    = getDataCountry()
dataGlobal = getDataCountry("Global") 
```

```python
import matplotlib.pyplot as plt
```

```python
dataMEX.columns
```

```python
#Total

for column in dataMEX.columns:
    fig , ax= plt.subplots()
    ax.plot(dataMEX[column])
    ax.plot(dataGlobal[column])
    ax.set_title(column)
    plt.show(fig)
    plt.close(fig)
```
