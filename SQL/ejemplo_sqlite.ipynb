### WHERE

> SELECT col \
> FROM tabla \
> WHERE condiciones;

Operadores de comparación: **< , >=, =, !=**

Operadores lógicos: **AND, OR, NOT**  (para multiples condiciones)

```python
pd.read_sql("""SELECT player_name, weight, height
                FROM Player
                WHERE height <= 180
                LIMIT 5;""", conn)
```

```python

pd.read_sql("""SELECT player_name, weight, height
                FROM Player
                WHERE height <= 180 AND weight > 190;""", conn)
```

```python
pd.read_sql("""SELECT player_name, weight, height
                FROM Player
                WHERE height <= 180 AND weight > 190 AND player_name != 'David Dunn';""", conn)
```

```python
pd.read_sql("""SELECT player_name, weight, height
                FROM Player
                WHERE height BETWEEN 180 AND 190
                LIMIT 5;""", conn)
```

```python
# NOT BETWEEN
pd.read_sql("""SELECT player_name, weight, height
                FROM Player
                WHERE height NOT BETWEEN 180 AND 190
                LIMIT 5;""", conn)
```

```python
pd.read_sql("""SELECT *
                FROM Match
                WHERE date BETWEEN '2014-03-15' AND '2014-04-15'""", conn)
```

### Unión de tablas: *JOIN*

* INNER JOIN : coincidencias entre las dos tablas
* LEFT JOIN : primera tabla completa, coincidencias con la segunda tabla, valores nulos
* RIGHT JOIN : segunda tabla completa, coincidencias con la primera tabla, valores nulos

> SELECT * FROM tabla1 \
> INNER JOIN tabla2 \
> ON tabla1.col_t1 = tabla2.col_t2

```python
pd.read_sql("""SELECT * FROM League
                INNER JOIN Country ON Country.id = League.country_id;""", conn)
```

```python
pd.read_sql("""SELECT player_name, penalties 
                FROM Player
                INNER JOIN Player_Attributes
                ON Player.player_api_id = Player_Attributes.player_api_id;""", conn)
```

```python
# si el nombre de la columna existe en ambas tablas, hay que especificar una tabla

pd.read_sql("""SELECT player_name, penalties, Player.player_api_id 
                FROM Player
                INNER JOIN Player_Attributes
                ON Player.player_api_id = Player_Attributes.player_api_id;""", conn)
```

```python
pd.read_sql("""SELECT player_name, SUM(penalties) AS total_penalties
                FROM Player
                INNER JOIN Player_Attributes
                ON Player.player_api_id = Player_Attributes.player_api_id
                GROUP BY player_name;""", conn)
```

```python
pd.read_sql("""SELECT player_name, SUM(penalties) AS total_penalties
                FROM Player
                INNER JOIN Player_Attributes
                ON Player.player_api_id = Player_Attributes.player_api_id
                GROUP BY player_name
                ORDER BY total_penalties DESC;""", conn)
```

```python
pd.read_sql("""SELECT player_name AS NOMBRE, SUM(penalties) AS total_penalties
                FROM Player
                INNER JOIN Player_Attributes
                ON Player.player_api_id = Player_Attributes.player_api_id
                GROUP BY NOMBRE
                ORDER BY total_penalties DESC;""", conn)
```

```python
pd.read_sql("""SELECT player_name AS NOMBRE, SUM(penalties) AS total_penalties
                FROM Player
                INNER JOIN Player_Attributes
                ON Player.player_api_id = Player_Attributes.player_api_id
                WHERE weight >= 180
                GROUP BY NOMBRE
                ORDER BY total_penalties DESC;""", conn)
```

```python
pd.read_sql("""SELECT Match.id, 
                    Country.name AS country_name, 
                    League.name AS league_name, 
                    season, 
                    stage, 
                    date,
                    HT.team_long_name AS  home_team,
                    AT.team_long_name AS away_team,
                    home_team_goal, 
                    away_team_goal                                        
                FROM Match
                JOIN Country on Country.id = Match.country_id
                JOIN League on League.id = Match.league_id
                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                WHERE country_name = 'Spain'
                ORDER by date
                LIMIT 10;""", conn)
```
