```python
import requests
import pandas as pd
```

```python
buscar_en_titulo = 'astro'
 
endpoint = 'https://es.wikipedia.org/w/rest.php/v1/search/title'

params = {
            'q' : buscar_en_titulo,
            'limit': 20
        }

request_wiki = requests.get(endpoint, params=params)

request_wiki.status_code
```

```python
df = pd.json_normalize(request_wiki.json()['pages'])
df
```

```python
for index in df.index:
  row = df.loc[index]
  buscar_titulo = row["key"]
  endpoint = 'https://es.wikipedia.org/w/api.php'
  params.update({'titles': buscar_titulo})
  requests_wiki = requests.get(endpoint, params=params)
  df_data = pd.json_normalize(requests_wiki.json()['query'])
  resume = df_data[df_data.columns[-1]][0]
  isscientific = int(("física" in resume.lower() or "ciencia" in resume.lower())) 
  df_pages.loc[len(df_pages.index)] = [row["id"], row["title"], resume, isscientific]

df_pages
```
